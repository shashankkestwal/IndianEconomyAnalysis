## Data cleaning and pre-processing

### Handling missing values

*Checking for missing or Nan values*

```python
sgdp.isna().sum()
```

*Missing value percentage in the dataset*

```python
sgdp.isna().sum()/len(sgdp)*100
```

*Dropping attribute with 25% missing or Nan values from the dataset*

```python
# total of 11 rows so each state must contain more than 25% of the data for furthur details
half = len(sgdp) / 4
x = sgdp.isna().sum() > half
drop_indexes = sgdp.columns[x]
# dropping columns from the dataframe 
sgdp.drop(drop_indexes, axis=1, inplace=True)
sgdp.shape
```

### Data Transformation

```python
sgdp = sgdp.set_index("State/Union Territory", drop = True)
```

## Data analysis and Visulisation

*Computing percentage change in growth in every consecutive year*

```python
percent_change = sgdp.pct_change()
percent_change = percent_change*(100)
percent_change
```

*Bar graph depecting average % growth of every state from 2011-2020*

```python
percent_change.mean().round(2).sort_values().plot.bar(title = "% Growth of each state", 
                                                      xlabel = "States", 
                                                      ylabel = "% change")
```

*Compairing the GDP growth rates of my home state(Uttarakhand, Himanchal Pradesh) with that of INDIA*

```python
#calulating all india GDP
sgdp["India"] = sgdp.sum(axis = 1)  
sgdp.head()
```

```python
perChange = sgdp.pct_change()*100
home_state = perChange[["Himachal Pradesh", 'Uttarakhand', 'India']]
comapred_states = home_state.sum()/ home_state.count()
comapred_states.plot(kind='bar',stacked=True, colormap = 'Dark2')
plt.title("Avg. % Growth of Home State vs National Avg. for Duration 2011-20", fontweight = 'bold')
plt.xlabel("States", fontweight = 'bold')
plt.ylabel("% growth rate", fontweight = 'bold')
```

Our Home states have relatively higher growth rates as compared India. 

*Plotting the GDP of all states of last year (2020- 2021)*

```python
result = sgdp.loc["2019-20"].copy()
result.drop(labels='India', inplace=True)
result.sort_values(inplace=True)
result.plot(kind='bar',stacked=True, title = "State Wise GDP in 2019-20", ylabel = "GDP (Lakh Crore)")
```

### <center> Overvation</center>
States with larger area and populations have high GDP as compared to those with smaller states and UT <br><br>
Maharashtra , TamilNadu and Gujarat have maximum shares in the total GDP of the country

### Relationship between per capita income vs Literacy rate

*Reading literacy data from file and transforming data to visualise*

```python
literacy_rate = pd.read_excel("literacy_rate.XLSX", sheet_name=0, skiprows=3, skipfooter=4) #reading data from excel
literacy_rate = literacy_rate.replace('-', np.nan)   #replacing empty cell with nan values
literacy_rate = literacy_rate.drop('Unnamed: 0', axis=1)    #dropping invalid attributes
literacy_rate = literacy_rate.set_index("State/Union Territory", drop = True)    #resting index 
literacy_rate = literacy_rate.transpose()       #transposing data
literacy_rate.rename(columns = {"ALL INDIA": "India"}, inplace = True)  #renaming columns 
literacy_rate
```

*Reading per capita income date from file and transforming data to visualise*

```python
perCapita = pd.read_excel("per_capita_income.XLSX", sheet_name=2, skiprows=5, skipfooter=2) #reading data from excel
perCapita = perCapita.replace('-', np.nan)  #replacing empty cell with nan values
perCapita = perCapita.drop('Unnamed: 0', axis=1)   #dropping invalid attributes
perCapita.rename(columns = {"Unnamed: 1" : "State/Union Territory"}, inplace = True)  #renaming index 
perCapita = perCapita.set_index("State/Union Territory", drop = True)  #resting index 
perCapita = perCapita.transpose()   #transposing data
perCapita
```

### Relationship between Inflation vs % GDP growth 

*Reading inflation data from file and transforming data to visualise*

```python
inflation = pd.read_excel("inflation_general.XLSX", skiprows = 3, skipfooter = 3) #reading data from excel
inflation.drop("Unnamed: 0", inplace = True, axis = 1)   #dropping invalid attributes
inflation.set_index("State/Union Territory", inplace = True)  #ressting index
inflation = inflation.transpose()  #transposing data
inflation.rename(columns = {"All India" : "India"}, inplace = True)  #renamming entries
inflation
```

```python
perChange.index
```

```python
inflation.index
```

*Plotting line chart depicting the relationship between GDP and Inflation*

```python
plt.plot(inflation.index[:len(inflation.index)-1], inflation["India"].iloc[:len(inflation.index)-1], label = "Inflation")
plt.plot(perChange.index[3:], perChange["India"].iloc[3:], label = "GDP growth")
plt.xlabel("Years")
plt.ylabel("% Change")
plt.legend()
plt.show()
```

### <center>Observation</center>
Times when the inflation rates of the country were going down the total GPD of the country was having a boom. <br> <br>
When the inflation rates continue to rise the GDP growth rates started to go down.
